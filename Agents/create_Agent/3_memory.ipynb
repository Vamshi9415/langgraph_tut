```python
from langchain.agents import create_agent
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(model = "gemini-2.5-flash")

agent = create_agent(
    model = model
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content = "My name is hitman and my favourite colour is blue")

response = agent.invoke({
    'messages' : [question]
})
response
```

```python
from pprint import pprint

pprint(response)
```

```python
question2 = HumanMessage(content = "whats my name and my favourite colour")

response2 = agent.invoke({
    'messages' : [question2]
})

response2
```

so we need to have short term memory to have an conversation so we use checkpoint memory from langgraph

```python
from langgraph.checkpoint.memory import InMemorySaver
#checkpointer is a way for persisting and managing the state os the graph like an agent across multiple executions
# without checkpoint graphs are stateless, checkpointer adds memory , on the next run it loads the saved state, so the graph remembers previous steps
# it uses thread_id to isolate states for different users 
# MemorySaver stores in RAM short term memory
agent = create_agent(
    model = model,
    checkpointer = InMemorySaver()
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content = " my name is hitman and my fav colour is blue")

config = {
    "configurable" : {
        "thread_id" : "1"
    }
}

response = agent.invoke(
    {'messages':[question]},
    config
)

pprint(response)
```

```python
question = HumanMessage(content = " whats my name and whats my fav colour")

response = agent.invoke(
    {'messages' : [question]},config
)

pprint(response)
```
