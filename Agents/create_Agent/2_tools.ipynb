```python
from langchain.tools import tool

@tool

def square_root(x : float) -> float :
    """ Calculate square root of the number """
    
    return x**0.5
```

```python
@tool("square_root",description= "calculate square root of a number")
def tool1(x: float) -> float :
    return x**0.5
```

```python
tool1.invoke({'x':121})
```

Adding to agents

```python
from langchain.agents import create_agent
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(model = "gemini-2.5-flash")
agent = create_agent(
    model = model,
    tools = [tool1],
    system_prompt= "you are maths teacher, use the tools to calculate the square root of the number"
)


```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="what is the square root of 121")

response = agent.invoke(
    {'messages' : [question]}
)

response['messages']
```

```python
from typing import Dict, Any
from tavily import TavilyClient
from langchain.tools import tool
tavily_client = TavilyClient() 

@tool
def websearch(query: str) -> Dict[str, Any]:
    """Search the web for information."""
    return tavily_client.search(query=query)  

# Test
result = websearch.invoke("who is the current captain of indian cricket team")
print(result) 
```

```python
from pprint import pprint

pprint(result)
```
